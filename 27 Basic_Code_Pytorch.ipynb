```python
#import statements
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

import torch 
import torch.nn as nn
```

```python
#parameters

RANDOM_STATE = 24
np.random.seed(RANDOM_STATE)  #setting random state to numpy
torch.manual_seed(RANDOM_STATE)   

EPOCHS = 2000  #no of epochs

TEST_SIZE = 0.2
```

```python
#reading dataset
data_df = pd.read_csv("wine.csv", header=None)
data_df.head()
```

```python
#splitting data for y
labels = data_df[data_df.columns[-1]]
labels
```

```python
#splitting data for X 
features_df = data_df.drop(data_df.columns[-1], axis = 1)
features_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_df, labels, 
                                                    stratify=labels, 
                                                    test_size=TEST_SIZE,
                                                    random_state=RANDOM_STATE)

X_train.shape, X_test.shape, y_train.shape, y_test.shape   #checking shapes
```

```python
print(type(X_train))
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)  #standardscaler transforms dataframe to numpyarray so no need to use to_numpy
X_test = sc.transform(X_test)

#label encoding
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

type(X_train), type(X_test), type(y_train), type(y_test)

#if label encoding is done dataframe gets transformed to numpyarray otherwise do this
#y_train = y_train.to_numpy()         #converting from dataframe to numpyarray using to_numpy
#y_test = y_test.to_numpy()
```

### preparing model

```python
X_train.shape[0]
```

```python
X_train.shape[1]   
```

```python
#model creation 

model = nn.Sequential(
    #nn.Linear(in_features =X_train.shape[1], out_features = 8)
    nn.Linear(X_train.shape[1], 8),          #shape of features
    nn.ReLU(),                               #activation
    nn.Linear(8,3)                           # 8 -> neurons of hidden layer, 3 -> output
)

model
```

```python
#to calculate loss
loss_fn = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters())
```

```python
type(X_train)
```

```python
#changing numpyarray to tensor 

train_X = torch.tensor(X_train, dtype=torch.float32)
train_Y = torch.tensor(y_train, dtype=torch.int64)

test_X = torch.tensor(X_test, dtype=torch.float32)
test_Y = torch.tensor(y_test, dtype=torch.int64)  #labels always int64 
```

```python
type(train_X)
```

```python
#for training model (train and test) and also plotting loss curve

loss_o, loss_t = [], []
for epoch in range(EPOCHS): #EPOCHS 

    #on training
    model.train()           #fitting the model
    outputs = model(train_X)
    loss = loss_fn(outputs, train_Y)
    loss_o.append(loss.item())
   
    #backpropagation of tain
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    ###### till this is sufficient for training 

    #on testing data by turing off gradient descent
    with torch.no_grad():
        model.eval()
        outputs = model(test_X)
        tloss = loss_fn(outputs, test_Y)
        loss_t.append(tloss.item())

    print(f'Epoch: {epoch:2d} | Loss: {loss:.6f} | tLoss: {tloss:.6f}')
```

```python
loss_df = pd.DataFrame({'loss': loss_o, 'test_loss':loss_t})
loss_df.plot()
```
