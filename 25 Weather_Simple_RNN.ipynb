```python
time_step = 24
#creating index
y_idx = np.arange(time_step, tmp_df.shape[0], time_step)
y_idx.shape

y_df = tmp_df.iloc[y_idx][['datetime', 'temp']]
y_df.head()
```

```python
y_df.shape
```

```python
y_df.shape[0]*time_step
```

```python
X_df = tmp_df.iloc[range(len(y_df)*time_step)]
X_df.shape
```

```python
X_temp = np.reshape(X_df['temp'].to_numpy(), 
               (y_df.shape[0], time_step))

X_hum = np.reshape(X_df['hum'].to_numpy(), 
               (y_df.shape[0], time_step))

X_viz = np.reshape(X_df['viz'].to_numpy(), 
               (y_df.shape[0], time_step))
```

```python
X_temp = X_temp[:, :23]
X_hum = X_hum[:, :23]
X_viz = X_viz[:, :23]

X_temp.shape, X_hum.shape, X_viz.shape
```

### Input Shape expected = '[batch, timestep, feature]'

```python
X_data = np.stack((X_temp, X_hum), axis = 2)
X_data.shape
```

```python
split = int(y_df.shape[0] * (1-TEST_SIZE))
X_train = X_data[:split]
X_test = X_data[split:]

y_train = y_df['temp'].to_numpy()[:split]
y_test = y_df['temp'].to_numpy()[split:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#model

h_units = 256    # hidden units   
input_shape=(time_step -1, X_train.shape[2])

model = tf.keras.Sequential()
model.add(tf.keras.layers.SimpleRNN(
    units= h_units,
    input_shape=input_shape,
    activation= 'tanh'))

model.add(tf.keras.layers.Dense(1, activation='linear'))

model.compile(loss = 'mean_squared_error', metrics = [tf.keras.metrics.RootMeanSquaredError()], optimizer = 'adam')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, 
                   epochs = EPOCHS, 
                   batch_size = BATCH_SIZE,
                   validation_data = [X_test, y_test],
                   verbose = 2)
```

```python
loss_df = pd.DataFrame(history.history)
fn_plot_tf_hist(loss_df)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
y_pred = np.append(y_train_pred, y_test_pred)

y_pred.shape
```

```python
res_df = y_df.copy()
res_df['pred'] = y_pred
res_df['datetime'] = res_df['datetime'].dt.date
res_df.head()
```
